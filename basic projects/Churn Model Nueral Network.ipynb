```python
import pandas as pd
```

```python
data = pd.read_csv("churn_model.csv")
```

```python
X = data.iloc[:, 3:13]
```

```python
Y = data.iloc[:, 13]
```

```python
X.head()
```

```python
gender = pd.get_dummies(X['Gender'],drop_first=True)
```

```python
gender
```

```python
geography = pd.get_dummies(X["Geography"], drop_first=True)
```

```python
geography
```

```python
X = pd.concat([X, geography, gender], axis=1)
```

```python
X.head()
```

```python
X = X.drop(["Geography", "Gender"], axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LeakyReLU,PReLU,ELU
```

```python
model2 = Sequential()
```

```python
model2.add(Dense(units=100, kernel_initializer='he_uniform', activation='relu', input_dim=11))
```

```python
model2.add(Dense(units=75, kernel_initializer='he_uniform', activation='relu'))
```

```python
model2.add(Dense(units=1, kernel_initializer='glorot_uniform', activation = 'sigmoid'))
```

```python
model2.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
classifier = model2.fit(x_train, y_train, validation_split=0.33, batch_size = 10, epochs = 100)
```

```python
y_pred = model2.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
```
