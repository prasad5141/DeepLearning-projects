```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('me.jpeg')
plt.imshow(img)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_grey = cv2.imread('me.jpeg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img_grey, cmap='gray')
```

```python
## Resizing Image
img_grey.shape
```

```python
resized_img = cv2.resize(img_grey, (420,640))
plt.imshow(resized_img)
```

### Flipping Image

```python
flip_image = cv2.flip(img_grey, 0)
plt.imshow(flip_image)
```

```python
flip_image = cv2.flip(img_grey, 1)
plt.imshow(flip_image)
```

```python
flip_image = cv2.flip(img_grey, -1)
plt.imshow(flip_image)
```

### Writing on Images

```python
import numpy as np

blank_image = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(blank_image, cmap='gray')
```

```python
blank_image.shape
```

```python
cv2.rectangle(img=blank_image, pt1=(400,0), pt2=(512,100), color=(255, 0,0), thickness=10)
plt.imshow(blank_image)
```

```python
cv2.circle(img=blank_image, center=(200,300), radius=50, color=(0,255,0), thickness=10)
plt.imshow(blank_image)
```

```python
cv2.circle(img=blank_image, center=(400,300), radius=50, color=(0,255,0), thickness=-1)
plt.imshow(blank_image)
```
