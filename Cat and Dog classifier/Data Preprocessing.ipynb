### Importing necessary modules

```python
import os
import shutil
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from torchvision import datasets,transforms

%matplotlib inline
```

### Loading data from D drive
I had downloaded Dogs vs Cats Dataset from Kaggle and stored this data outside of my project, i don't want to include train&test data in my project folder

```python
data_dir = 'D:\cat_dog_data'  #Data location

files = os.listdir(data_dir)
print(files)   
# I have two sub-directories in side cat_dog_data folder where 1 represents dog and 0 represents cats.
# When i download dataset from kaggel both cats and dogs images are mixed we have separate those images by using their names
                
```

```python
# Initially i am keeping all images in '0' sub-folder so i have to take all dog images than move to 1
all_images_dir = os.path.join(data_dir, files[0])

all_image_files = os.listdir(all_images_dir)
```

```python
len(all_image_files)
```

```python
### Running loop over all images than find dog images, move those dog images to folder 1
destination_dir = os.path.join(data_dir, '1')

i = 1
for image in all_image_files:
    name = image.split('.')[0]   #taking image name
    if name == 'dog':
        image_path = os.path.join(all_images_dir, image) # if image name is equal to dog than moving that file to '1' folder
        shutil.move(image_path, destination_dir)
```

### LIsting all dog images

```python
cat_dir = os.path.join(data_dir, '0')
dog_dir = os.path.join(data_dir, '1')

cat_images = os.listdir(cat_dir)
dog_images = os.listdir(dog_dir)

for i in range(10):
    dog_image = dog_images[i]
    
    dog_image_path = os.path.join(dog_dir, dog_image)
    
    dog_image = mpimg.imread(dog_image_path)
    
    plt.show()
    plt.imshow(dog_image)
```

### LIsting all cat images

```python
for i in range(10):
    cat_image = cat_images[i]
    
    cat_image_path = os.path.join(cat_dir, cat_image)
    
    cat_image = mpimg.imread(cat_image_path)
    
    plt.show()
    plt.imshow(cat_image)
```
